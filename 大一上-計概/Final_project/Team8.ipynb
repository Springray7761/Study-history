```python
%%writefile Final_Project.cpp
#include <iostream>
#include <string>
#include <algorithm>
#include <ctime>
#include <random>
#include <set>
using namespace std;

string generate_secret_number(int digit_nums,int digit_forawrd) {/*生產答案*/
    string numbers = "";
    if (digit_forawrd>36&&digit_forawrd<=62){
        for(int i=0;i<=digit_forawrd-37;i++) numbers+=char('A'+i);
        digit_forawrd=36;
    }
    if(digit_forawrd>10&&digit_forawrd<=36){
        for(int i=0;i<=digit_forawrd-11;i++) numbers+=char('a'+i);
        digit_forawrd=10;
    }
    if(digit_forawrd>=2&&digit_forawrd<=10)
    for(int i=0;i<digit_forawrd;i++)
    numbers+=char('0'+i);
    mt19937 rng(random_device{}());
    shuffle(numbers.begin(),numbers.end(),rng);
    return numbers.substr(0, digit_nums);
}

pair<int, int> bulls_and_cows(const  string& secret, const  string& guess) {/*檢測bulls and cows*/
    int bulls = 0, cows = 0;
    for (int i=0;i<guess.size();++i) {
        if (guess[i]==secret[i]) {
            bulls++;
        } else
        if (secret.find(guess[i]) != string::npos){
            cows++;
        }
    }
    return make_pair(bulls, cows);
}
int real_number(char digit){/*把代表的字母轉換成數字*/
    int temp=digit-'0';
    if(temp>=0&&temp<=9) return temp;
    if(temp>=49&&temp<=74) return temp-39;
    if(temp>=17&&temp<=42) return temp+19;
    return 0;
}
bool is_valid_guess(const string & guess) {/*檢測正確輸入*/
    set<char> unique_chars(guess.begin(), guess.end());
    return unique_chars.size() == guess.size();
}

int main() {
    srand(static_cast<unsigned int>(time(0)));
    vector<string> secret_number;
    string guess;
    bool has_been_changed=false;                    /*各參數前置*/
    int dice_temp=0,digit_nums[2]={0},mode_sel,player_nums=1,attempts=0,again=1,playersetnumber=0,dice_num=0,settings=0,digit_forawrd=10;
    int score_30[2]={0,0};
    vector<int> score;
    cout<<"\"Welcome to bulls and cows. After 14 days of development, hopefully it would have been worth the wait. Thanks and have fun\" -not Gabe Newell"<<endl<<"Six modes available now:"<<endl;
    cout<<"1.Single Player Mode"<<endl<<"2.Competition Mode"<<endl<<"3.pk mode"<<endl<<"4.決戰30 mode"<<endl<<"5.Story Mode(Closed Beta)"<<endl<<"6.Online Mode(Closed Beta)"<<endl;
    while(1){                                       /*模式介紹*/
        cout<<"Which mode you want to play(Just type the number of the mode,type in alphabet or float to quit)?"<<endl;
        cin>>mode_sel;                                      /*模式輸入*/
        if(mode_sel==1||mode_sel==3) break;
        if(mode_sel==2){
            cout<<"How many players?"<<endl;
            cin>>player_nums;
            cout<<"Player can set answer or not?(0 as No,1 as Yes)"<<endl;
            cin>>playersetnumber;
            break;
        }
        if(mode_sel==4) {
            playersetnumber=2;
            break;
        }
        if(typeid(mode_sel).name()!="int") return 0;
        cout<<"Error input,Please try again."<<endl;
    }
    if(mode_sel==1||mode_sel==2||mode_sel==3) cout<<"We featured a top limit of 62 digits.(\'a\' as 10,\'A\'as 37 etc.)"<<endl;/*各模式前置調整、決勝30介紹*/
    if(mode_sel==4) {
        cout<<"In 決戰 30,you will set the answer for your opponent."<<endl;
        cout<<"Once you get the right answer ,you will get the sum of your answer as the points you get."<<endl;
        cout<<"When you have your points 30 or over 30,you will win."<<endl;
    }
    while(again){
        if(mode_sel!=4){
            cout<<"How many digits you want(Enter 0 to quit)?"<<endl;      /*答案型式輸入*/
            cin>>digit_nums[0];
            if(digit_nums[0]==0) break;
            cout<<"How much as a forawrd(多少進一位,we provide the 62 as a forward for the highest,and 2 as the least.)?"<<endl;
            cin>>digit_forawrd;
        }
        for(int i=0;(i<player_nums||(mode_sel==4&&score_30[0]<30&&score_30[1]<30));i++){
            for(int i=0;i<playersetnumber;i++){                                                                 /*答案輸入或生成、決勝30介紹*/
                if(mode_sel==4&&i==0){
                    cout<<"The answer in 決勝30 will be limit to 10 numbers,and you must talk to your opponent to decide how many digits to play.(We recommend 3)"<<endl;
                    cout<<"Player A:"<<endl;
                }
                else if(mode_sel==4) cout<<"Player B:"<<endl;
                cout<<"Please type in the answer(because system(\"cls\") isn't working properly.We use cout 50 lines of nothing as a alternative.)"<<endl;
                string answer_temp;
                set<char> unique_chars;
                do{
                    cin>>answer_temp;
                    if(mode_sel==4) digit_nums[i]=answer_temp.size();
                    if(!is_valid_guess(answer_temp)||(i==1&&(digit_nums[1]!=digit_nums[0]))) cout<<"Error input,please try again."<<endl;
                }while(!is_valid_guess(answer_temp)||(i==1&&(digit_nums[1]!=digit_nums[0])));
                system("clear");   /*for(int i=0;i<50;i++) cout<<endl;  */
                secret_number.push_back(answer_temp);
                answer_temp="";
            }
            if(playersetnumber==0&&mode_sel!=4) secret_number.push_back(generate_secret_number(digit_nums[0],digit_forawrd));
            if(mode_sel==3){
                secret_number.push_back(generate_secret_number(digit_nums[0],digit_forawrd));
                digit_nums[1]=digit_nums[0];
                cout<<"In PK mode,there will be lots of dice rolling."<<endl;
                cout<<"Once the rounds are over 3."<<endl;
                cout<<"In each round,you will roll a dice."<<endl;                              /*PK mode 介紹*/
                cout<<"You have the a 50 50 chance to either roll the effect of dice,or triger the effect of the dice."<<endl;
                cout<<"So there is a chance of you keeping rolling the effect of the dice,but not having the effect affect."<<endl;
                cout<<"And we will left the effect for you to find out."<<endl;
            }
            while (true) {
                if(mode_sel==2) cout<<"Player "<<i+1<<" ";
                if(attempts==6&&mode_sel==3) cout<<"The dice event has started."<<endl;             /*PK mode 骰子部分*/
                if((mode_sel==3||mode_sel==4)&&attempts%2==0) cout<<"Player A:"<<endl;
                else if ((mode_sel==3||mode_sel==4)&&attempts%2==1) cout<<"Player B:"<<endl;
                if(attempts>=6&&mode_sel==3){
                    if(rand()%2==0){
                        cout<<"The dice have been rolled,the effect has changed."<<endl;
                        do{
                            if(has_been_changed) dice_num=rand()%5;                             /*骰子效果變換*/
                            else dice_num=rand()%6;
                        }while(dice_num==dice_temp);
                        dice_temp=dice_num;
                    }
                    else{
                        if(dice_num==0){    /*骰子效果1 揭示兩個字 並提示其中一個的位置*/
                            int change_first=rand()%digit_nums[attempts%2],change_second=0;
                            cout<<"One of the digit in the answer is "<<secret_number[attempts%2][change_first]<<"."<<endl;
                            do{
                                change_second=rand()%digit_nums[attempts%2];
                            }while(change_first==change_second);
                            cout<<"The number "<<change_second+1<<" digit in the answer is "<<secret_number[attempts%2][change_second]<<"."<<endl;
                        }
                        if(dice_num==1){
                            int odds=0,evens=0;         /*骰子效果2  提示答案中奇偶數*/
                            for(int counter=0;counter<secret_number[attempts%2].size();counter++){
                            if(real_number(secret_number[attempts%2+1][counter])%2==0) evens++;
                            else odds++;
                            }
                            cout<<"You have "<<odds<<" odds in the answer."<<endl;
                            cout<<"You have "<<evens<<" evens in the answer."<<endl;
                        }
                        if(dice_num==2){
                            int the_add=0;        /*骰子效果3  對手位數增加1*/
                            digit_nums[attempts%2+1]++;
                            if(digit_nums[attempts%2+1]<=10){
                                do{
                                    the_add=rand()%10;
                                }while(secret_number[attempts%2+1].find(to_string(the_add))!=string::npos);
                                secret_number[attempts%2+1]+=to_string(the_add);
                            }
                            if(digit_nums[attempts%2+1]>10&&digit_nums[attempts%2+1]<=36) secret_number[attempts%2+1]+=to_string('a'+digit_nums[attempts%2+1]-11);
                            if(digit_nums[attempts%2+1]>36&&digit_nums[attempts%2+1]<=62) secret_number[attempts%2+1]+=to_string('A'+digit_nums[attempts%2+1]-37);
                            cout<<"The number of the digit of your enemy's answer has been plus 1,and the change will be on the last number of the answer."<<endl;
                        }
                        if(dice_num==3){  /*骰子效果4 提示答案各數字加起來的值*/
                            int total=0;
                            for(int temp=0;temp<secret_number[attempts%2].size();temp++) total+=real_number(secret_number[attempts%2][temp]);
                            cout<<"The sum of your answer is "<<total<<"."<<endl;
                        }
                        if(dice_num==4){   /*骰子效果5 隨機變換對方答案中的其中一個數字*/
                            if(digit_nums[attempts%2+1]>10) cout<<"The answer has reached its limit,notthing happend."<<endl;
                            else{
                                int choose_num =rand()%digit_nums[attempts%2+1],temp2=0;
                                do{
                                    temp2=rand()%digit_forawrd;
                                }while(secret_number[attempts%2+1].find(to_string(temp2))!=string::npos);
                                secret_number[attempts%2+1][choose_num]=temp2;
                                cout<<"The number "<<choose_num+1<<" digit of your enemy has changed to something else."<<endl;
                            }
                        }
                        if(dice_num==5){   /*骰子效果6 重洗對方的牌*/
                            mt19937 rng(random_device{}());
                            shuffle(secret_number[attempts%2+1].begin(),secret_number[attempts%2+1].end(),rng);
                            cout<<"The answer of your enemy has beed shuffled,and this will only happend once."<<endl;
                            has_been_changed=true;
                        }
                    }
                }
                if(mode_sel==4) settings=(attempts+1)%2;                   /*輸入與檢測*/
                else if(mode_sel==3) settings=attempts%2;
                else  settings=0;
                cout << "Enter your guess ( "<<digit_nums[settings]<<" unique digits,enter 0 to quit): ";
                cin >> guess;
                if(guess=="0") return 0;
                if(!is_valid_guess(guess)||guess.size()!=digit_nums[settings]){
                    cout << "Invalid input. Please try again.\n";
                    continue;
                }/*輸入與答案判定*/
                auto [bulls, cows] = bulls_and_cows(secret_number[settings], guess);
                attempts++;
                cout << bulls << " Bulls, " << cows << " Cows\n";
                if (bulls==digit_nums[settings]) {
                    if(mode_sel==1){
                        cout << "Congratulations! You have guessed the number in " << attempts << " attempts.\n";           /*答對輸出結果*/
                        secret_number.clear();
                        attempts=0;
                        break;
                    }
                    if(mode_sel==2){
                        cout << "Congratulations! Player "<<i+1<<" guessed the number in " << attempts << " attempts.\n";   /*答對輸出結果*/
                        score.push_back(attempts);
                        attempts=0;
                        secret_number.clear();
                        break;
                    }
                    if(mode_sel==3){                                        /*答對輸出結果*/
                        cout << "Congratulations! Player ";
                        if(settings==0){
                            cout<<"A";
                            attempts++;
                        }
                        else cout<<"B";
                        cout<<" guessed the number in "<<attempts/2<< " attempts,and won.\n";
                        secret_number.clear();
                        has_been_changed=false;
                        attempts=0;
                        break;
                    }
                    if(mode_sel==4){
                        cout << "Congratulations! Player ";                         /*答對輸出結果*/
                        if(settings==1) cout<<"A";
                        else cout<<"B";
                        int total=0;
                        for(int temp=0;temp<secret_number[(attempts+1)%2].size();temp++) total+=real_number(secret_number[(attempts+1)%2][temp]);
                        score_30[settings]+=total;
                        cout<<" guessed the number ,and won "<<total<<" points,now has "<<score_30[settings]<<" points."<<endl;
                        attempts=0;
                        settings=0;
                        secret_number.clear();
                        break;
                    }
                }
            }
        }
        if(mode_sel==4){
            cout<<"Player ";
            if(settings==0)cout<<"A";
            else cout<<"B";                                     /*成績總結*/
            cout<<" has won the game."<<endl;
            score_30[0]=0;
            score_30[1]=0;
        }
        if(mode_sel==2){
            vector<int> score_temp = score;
            sort(score.begin(),score.end());
            cout<<"The game has end."<<endl;
            cout<<"Final score:"<<endl;                                     /*成績總結*/
            for(int j=0;j<player_nums;j++){
                if(score[j]==score[j+1]){/*not yet test*/
                    cout<<"Player "<<find(score_temp.begin(),score_temp.end(),score[j])-score_temp.begin()+1<<" had "<<score[j]<<" tries,end up with "<<j+1<<" place."<<endl;
                    score_temp[find(score_temp.begin(),score_temp.end(),score[j])-score_temp.begin()]=0;
                }
                else cout<<"Player "<<find(score_temp.begin(),score_temp.end(),score[j])-score_temp.begin()+1<<" had "<<score[j]<<" tries,end up with "<<j+1<<" place."<<endl;
            }
            score_temp.clear();
            score.clear();
        }
        cout << "Wanna play again?(Enter 0 as No,1 as Yes)";        /*重完詢問*/
        cin>>again;
    }
    return 0;
}
```
